# 2. Estatística descritiva

Aqui serão apresentados e discutidos **conceitos básicos de estatística**, que permitam a descrição de um conjunto de dados.

Assim, você deverá ser capaz de:

* distinguir dado de informação,
* identificar as escalas de medidas de variáveis, 
* utilizar as medidas de tendência central e dispersão. 

Na estatística são utilizadas diversas medidas que caracterizam a distribuição dos valores de uma determinada variável numérica. Essas medidas podem ser classificadas em medidas de tendência central, medidas de dispersão, medidas de assimetria e medidas de curtos. 

### 2.1 Medidas de tendência central

As medidas de tendência central são aquelas que buscam refletir o ponto de equilíbrio dos dados. Não há uma única medida de tendência central, e a razão para isso será mostrada com exemplos.

##### 2.1.1 Média




```r
#Para obtermos a média de uma variável x no R, utilizamos a função mean:
mean(x)
```

##### 2.1.2 Mediana

É definida como o valor tal que 50% dos valores da variável estão acima da mediana e 50% estão abaixo.

```r
#No R, temos a função median para fazer o cálculo da mediana. Para a variável x, o valor da mediana será obtido assim:
median(x)
```

##### 2.1.3 Moda

Medida de maior frequência em um conjunto de dados. Os passos para se obter
a moda são:
* encontrar todos os valores distintos da variável;
* obter a frequência de cada valor distinto;
* selecionar o valor (ou valores) com a maior frequência para obter a moda.

```r
#Para moda

#A função max(temp) retorna a frequência máxima em uma sequência temp
max(temp)
#A expressão temp == max(temp) vai retornar todos os itens de temp cujas contagens sejam iguais à frequência máxima. 
#Logo temp[temp == max(temp)] retorna os elementos da tabela de frequência com frequência máxima:
temp[temp == max(temp)]
#Finalmente names(temp)[temp == max(temp)] retorna os valores com frequência máxima.
```

#### Que medida de tendência central deve ser usada ou apresentada em relatório ou artigo científico?

Depende dos dados! Como vimos nos exemplos acima, podem ocorrer diversas situações:
* os dados estão distribuídos simetricamente em torno de um ponto central;
* existem valores no conjunto de dados que se desviam marcadamente dos valores típicos (conhecidos em inglês por “outliers”);
* as variáveis podem estar distribuídas de maneira assimétrica em maior ou menor grau;
* os dados podem estar distribuídos em torno de mais de um valor;
* etc.

No primeiro caso acima, a média é uma medida adequada. Já no terceiro caso, e às vezes no segundo, a mediana seria mais apropriada e, finalmente, no quarto caso, a moda seria a medida escolhida.

### 2.2 Medidas de dispersão

Medidas de tendência central ou localização dos dados não dão a visão completa dos mesmos. Para melhor interpretarmos os dados, também necessitamos saber como estes estão ‘espalhados’, isto é, se os dados estão localizados em sua maioria em torno da medida de tendência central ou estão mais dispersos.

##### 2.2.1 Amplitude

É a distância entre o maior e o menor valor do conjunto de dados. Usualmente, ela é calculada por `max(X) - min(X)`, onde X é a variável de interesse. 

```r
#A função range no R fornece o menor e o maior valor da variável entre parênteses:
range(x1); range(x2)

#A partir da função range, a amplitude de cada uma das variáveis poderia ser obtida, subtraindo-se o primeiro valor do segundo valor gerado pela função range. 
#A função diff calcula essa diferença. Podemos ver que a amplitude é a mesma para as variáveis x1 e x2
diff(range(x1))
diff(range(x2))
```
